# Initial Data Exploration

## Dataset description

In this assignment I will use [MovieLens 100K dataset](https://grouplens.org/datasets/movielens/100k/) consisting user ratings to movies.

* It consists of 100,000 ratings from 943 users on 1682 movies
* Ratings are ranged from 1 to 5
* Each user has rated at least 20 movies
* It contains simple demographic info for the users (age, gender, occupation, zip code)

```python
import pandas as pd
import zipfile
import matplotlib.pyplot as plt
```

```python
with zipfile.ZipFile('../data/raw/ml-100k.zip', 'r') as zip_ref:
        zip_ref.extractall('../data/raw/')
```

## Data

```python
ratings = pd.read_csv('../data/raw/ml-100k/u.data', sep='\t', header=None, 
                   names=['user_id', 'movie_id', 'rating', 'timestamp'])
```

```python
ratings
```

```python
ratings.info()
```

```python
# Distribution of ratings

plt.hist(x=ratings['rating'], bins=5, rwidth=0.8)
plt.xlabel('Rating')
plt.ylabel('Number of ratings')
plt.title('Distribution of ratings')
plt.show()
```

```python
# Number of ratings of each user

ratings['user_id'].value_counts()
```

```python
# Number of ratings of each movie

ratings['movie_id'].value_counts()
```

**Insights from the data:**

- Data is cleaned
- Each user has rated from 20 to 737 movies
- Movies have different number of ratings (from 1 to 583)
- Most of the ratings vary from 3 to 5

## Item

```python
movies = pd.read_csv('../data/raw/ml-100k/u.item', sep='|', encoding='latin-1',
                     names=['movie_id', 'movie_title', 'release_date', 'video_release_date', 
                          'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy',
                          'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
                          'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'])
```

```python
movies.head()
```

```python
movies.info()
```

**Insights from the data:**

- Video release date is null for all movies -> I will drop this column
- Genres represented in a convinient way and can be used for model trainings (as movie metadata)


## User

```python
user = pd.read_csv('../data/raw/ml-100k/u.user', sep='|',
                     names=['user_id', 'age', 'gender', 'occupation', 'zip_code'])
```

```python
user.head()
```

```python
user.info()
```

```python
len(user['age'].unique())
```

```python
# Distribution of user age 

plt.hist(x=user['age'], bins=61, rwidth=0.8)
plt.xlabel('Age')
plt.ylabel('Number of users')
plt.title('Age distribution')
plt.show()
```

```python
# User occupation

user['occupation'].value_counts().plot(kind='bar')
plt.title('Occupation of users')
plt.show()
```

**Insights from the data:**

- Age, gender, and occupation can be used in model training
- Gender should be turned into numerical data (for example, 0 for male and 1 for female)
- Occupation can be turned in the form line the movie genres (get_dummies)
- Zipcode containes letters, so it will be dropped

## Preprocess data

### Process user data

```python
user['gender'] = user['gender'].replace({'M': 0, 'F': 1})
user = pd.get_dummies(user, columns=['occupation'])
```

```python
user.head()
```

```python
# Merge all data into one dataset

data = pd.merge(ratings, movies, on='movie_id')
data = pd.merge(data, user, on='user_id')

# Drop unneeded columns
data = data.drop(columns=['timestamp', 'video_release_date', 'IMDb_URL', 'zip_code'])
```

```python
# Save preprocessed data

data.to_csv('../data/interim/data.csv', index=False)
```
